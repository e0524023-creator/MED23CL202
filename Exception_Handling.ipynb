```python
print("Demonstration of Common Python Exceptions\n")
try:
  print(10/0)
except ZeroDivisionError:
  print("X ZeroDivisionError:You cannot divide by zero.\n")

try:
  num = int("abc")
except ValueError:
  print("X valueError: Cannot convert string to integer.\n")

try:
  result = "2" + 2
except TypeError:
  print("X TypeError: You tried to add string and integer.\n")

try:
  lst = [1,2,3]
  print(lst[5])
except IndexError:
  print("X IndexError: List index out of range.\n")

try:
  my_dict = {'name': 'Anu'}
  print(my_dict['age'])
except KeyError:
  print("X FileNotFoundError: The file does not exist.\n")

try:
  print(undefined_variable)
except NameError:
  print("X NameError: Variable is not defined.\n")
```
