# **Dictionary Operations in Python**

C - Create

R - Read

U - Update

D - Delete


## **Create(Add new key-value pair)**

```python
student={}
student["name"]="Alice"
student["age"]=20
student["grade"]="A"
print(student)
```

## **Read**

```python
print(student["name"])
print(student.get("age"))
print(student.get("grade"))
print(student.get("grade","Not Available"))
```

## **Update**

```python
student={}
student["name"]="Bob"
student["age"]=19
student["grade"]="O"
print(student)
```

## **Delete**

```python
del student["grade"] #Completely removes key and value
print(student)
```

```python
removed=student.pop("age")
#pop will remove the key but the value will be present
#still cannot access the value
print(student)
print(removed)
```

## **Nested Dictionary**

```python
#Create a nested dictionary with student data
student={
    "Alice":{"Math":80,"Science":90,"English":95},
    "Bob":{"Math":75,"Science":100,"English":80},
    "Charlie":{"Math":65,"Science":78,"English":100},
    "David":{"Math":55,"Science":90,"English":50},
    "Eva":{"Math":72,"Science":78,"English":66}
}
print(student)
```

```python
import pandas as pd
data=pd.DataFrame(student)
data
```

```python
data=pd.DataFrame(student).T
data
```

```python
data["Average"]=data.mean(axis=1)
#axis=1 denoted row wise access to the dataframe
data
```

```python
data["Math"]=data.mean(axis=0)
#axis=0 denotes column wise access to the dataframe
data
#The result is not a number(NaN -not a number) because there is no variable to store the average
```

## **Data Visualizations**

```python
import matplotlib.pyplot as plt
#Bar Chart for students average performance
plt.figure(figsize=(8,5))
data["Average"].plot(kind="bar",color="skyblue",edgecolor="black")
plt.title("Students Average Performance")
plt.xlabel("Students")
plt.ylabel("Average Score")
plt.show()
```

## **How to visualize subjectwise trend**

```python
plt.figure(figsize=(8,5))
for subject in["Math","Science","English"]:
  plt.plot(data.index,data[subject],marker="o",label=subject)
  plt.title("Subjectwise Performance")
  plt.xlabel("Students")
  plt.ylabel("Score")
  plt.show()
```

```python
plt.figure(figsize=(8,5))
data[["Math","Science","English"]].plot(kind="box")
plt.title("Marks Distribution")
plt.xlabel("Subjects")
plt.ylabel("Score")
plt.show()
```
